# 1. Line Chart
A line chart shows how data changes over time by connecting points with a line.

It is mainly used to visualize trends and patterns.

```python
plt.figure(figsize=(5,3))  # figure size

# data 
x= [1,2,3,4,5]
y= [10,20,25,35,45]

plt.plot(x,y)  # line chart
plt.show()  # display chart in jupyter-notebook
```

# 2. Customize Chart
A customized chart is a graph where you modify elements like color, labels, title, size, and style to improve visualization and clarity.

```python
plt.figure(figsize=(5,3))  # figure size

plt.plot(x,y, color='green', marker='o', linestyle= 'dashed', linewidth=2, markersize=12)

plt.title('data title')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')

plt.grid()  # to show grid on page
plt.show()
```

# 3. Multiple Plot (Multiple Line & Legends)
A multiple line chart shows more than one data line on the same graph to compare trends.

Legends are used to identify each line clearly.

```python
plt.figure(figsize=(5,3))  # figure size

# data 
x= [1,2,3,4,5]
y= [10,20,25,35,45]
z= [20,30,35,45,55]

plt.plot(x,y,label='sales 2024') # ploting y data
plt.plot(x,z,label='sales 2024') # ploting z data

plt.title('YOY Sales')    #YOY- year on year
plt.xlabel('Months')
plt.ylabel('Sales')

plt.legend() # show legends
plt.show()
```

# 4. Bar Chart
A bar chart is used to compare values across different categories using rectangular bars.

```python
plt.figure(figsize=(5,3))  # figure size

# data 
x= ['A','B','C','D','E']
y= [10,20,25,35,45]

plt.bar(x,y)
plt.title('Bar Chart Example')
plt.show()
```

# 5. Countplot
A count plot shows the number of occurrences of each category using bars.

▫️Shows frequency of categorical data
▫️No need for manual counting (automatic)
▫️Created using Seaborn
▫️Supports **Hue** for sub-category comparison
▫️Useful in EDA (Exploratory Data Analysis)

Hue means adding a second categorical variable to divide data into sub-groups using different colors.

```python
plt.figure(figsize=(5,3))  # figure size

# data 
x= ['A','B','C','D','E']
y= [10,20,25,35,45]

plt.bar(x,y , color = ['g','k','r','y','b'])  # to show diff-diff colour
plt.title('Countplot Example')
plt.show()
```

# 6. Histogram
A histogram is a graph used to show the distribution of numerical (continuous) data by dividing it into intervals called bins.

▫️X-axis → Data intervals (bins)
▫️Y-axis → Frequency (count)
▫️Bars are connected (no gaps)
▫️Used to understand data distribution

```python
plt.figure(figsize=(5,3))  # figure size

#data
import random
data=[random.randint(1,10) for _ in range(100)]

plt.hist(data, bins=9)   # histogram chart
plt.title('Histogram Example')
plt.show()
```

# 7. Pie Chart
A pie chart is used to show proportions or percentages of a whole using circular slices.

▫️Each slice represents a category
▫️Size of slice = proportion/value
▫️Total = 100%
▫️Best for percentage distribution

```python
plt.figure(figsize=(4,4))  # figure size

# data 
categories= ['A','B','C','D','E']
sales= [10,20,55,35,45]

plt.pie(sales, labels=categories, autopct= '%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
plt.figure(figsize=(4,4))  # figure size

# data 
categories= ['A','B','C','D','E']
sales= [10,20,55,35,45]

plt.pie(sales, labels=categories, autopct= '%1.1f%%', explode=[0,0.1,0.2,0,0.1])     # explode - is used to separate all charts

plt.title(' Exploded Pie chart')
plt.show()
```

# 8. Scatter plot
Used to find relationship between variables

▫️Shows relationship between two variables (X & Y)
▫️Each point = one observation
▫️All points have same size
▫️Used to find correlation/trend

```python
plt.figure(figsize=(3,3))  # figure size

# data 
x= [1,2,3,4,5]
y= [10,20,25,35,45]

plt.scatter(x,y)
plt.title('Scatter Plot Example')
plt.show()
```

# 9. Bubble Plot
A bubble plot is a scatter plot where the size of each point represents a third variable.

▫️Extension of scatter plot
▫️Uses three variables (X, Y, Size)
▫️Bubble size represents extra data
▫️Used for multi-dimensional analysis

```python
plt.figure(figsize=(3,3))  # figure size

# data 
x= [1,2,3,4,5]
y= [10,20,25,35,45]

plt.scatter(x,y, color= 'y', edgecolors='k') # edgecolors for outerline
plt.title('Bubble Plot Example')
plt.show()
```

# 10. Box Plot
A box plot (also called a box-and-whisker plot) is a graphical way to summarize the distribution of data using five key values.

▫️Center (median)
▫️Spread (range & IQR)
▫️Skewness
▫️Outliers

```python
plt.figure(figsize=(5,3))  # figure size

# data 
x= [1,2,3,4,5]

plt.boxplot(x)
plt.title('Box Plot Example')
plt.show()
```

# 11. Subplot
A subplot is used to display multiple plots in a single figure.

▫️Divides figure into a grid (rows × columns)
▫️Each plot is placed in a specific position
▫️Helps compare multiple graphs together

```python
# data-1 - bar chart
categories = ['Mon','Tue','Wed','Thu','Fri']
sales = [10, 20, 55, 35, 45]

# data-2 - scatter plot
y1 = [10, 20, 25, 35, 45]
y2 = [20, 30, 35, 45, 55]

plt.figure(figsize=(10,3))  # figure size

# first plot- bar chart
plt.subplot(1,2,1)   # row, column, position
plt.bar(categories, sales)
plt.title("Daily Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")

# second plot- scatter chart
plt.subplot(1,2,2)   # row, column, position
plt.scatter(y1, y2)
plt.title("User Example")
plt.xlabel("User1")
plt.ylabel("User2")

plt.show()
```

# 12. Matplotlib with Pandas - real data

```python
# create df
import pandas as pd

# data
data = {
    'Month' : ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales' : [12000, 11000, 13000, 25000]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(6,3))  # fig. size

plt.bar(df['Month'],df['Sales'])
plt.title('Matplotlib with Pandas')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

# 13. Object-Oriented API
he Object-Oriented API in Matplotlib is a way of plotting where you control plots using objects like Figure and Axes instead of pyplot functions.

▫️Figure (fig) → Whole canvas (entire plot area)
▫️Axes (ax) → Individual plot (graph inside figure)

```python
# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axes (2 subplots)
fig, axs = plt.subplots(2, 1, figsize=(6,3))

# ---- First Plot (Sine) ----
axs[0].plot(x, y1, label='sin(x)', linewidth=2)
axs[0].set_title("Sine Wave")
axs[0].set_xlabel("X values")
axs[0].set_ylabel("sin(x)")
axs[0].legend()
axs[0].grid(True)

# Add annotation (arrow sign)
axs[0].annotate('Peak', xy=(1.57, 1), xytext=(3, 0.5),
                arrowprops=dict(facecolor='black', shrink=0.05))

# ---- Second Plot (Cosine) ----
axs[1].plot(x, y2, label='cos(x)', linestyle='--', linewidth=2)
axs[1].set_title("Cosine Wave")
axs[1].set_xlabel("X values")
axs[1].set_ylabel("cos(x)")
axs[1].legend()
axs[1].grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

# 14. 3D Plot

```python
fig = plt.figure(figsize=(4, 4))  # set figure size
ax = fig.add_subplot(111, projection='3d')  # 3D axes

x = np.random.rand(50)  # random x values
y = np.random.rand(50)  # random y values
z = np.random.rand(50)  # random z values

ax.scatter(x, y, z, c='red', marker='o')  # 3D scatter plot

ax.set_title("3D Scatter Plot")  # title

plt.show()  # display plot
```

```python
# Import 3D toolkit
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create figure and 3D axes
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and title
ax.set_title("3D Surface Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Show plot
plt.show()
```
