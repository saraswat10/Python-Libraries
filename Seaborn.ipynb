```python
# import seaborn
import seaborn as sns
# import matplotlib
import matplotlib.pyplot as plt
# import pandas
import pandas as pd
# import numpy as np
import numpy as np
```

```python
# import dataset from github
penguins= sns.load_dataset("penguins")
penguins.head(10)
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

# 1. Line plot
Shows trend over time or continuous data.

Points are connected with lines

```python
plt.figure(figsize=(6,3))
sns.set_style('whitegrid')  # used for themes
sns.lineplot(x="bill_length_mm",y= "bill_depth_mm",data=penguins, hue='island',style="sex")
plt.show()
```

# 2. Scatterplot
Shows relationship between two numerical variables.

Displays individual data points

```python
plt.figure(figsize=(6,3))
sns.set_style('darkgrid')  # used for background / themes

sns.scatterplot(x="body_mass_g", y="flipper_length_mm",data=penguins, hue= "island")
# hue is used to add an extra categorical variable to your plot using different colors.

# sns.scatterplot(x="body_mass_g", y="flipper_length_mm",data=penguins, hue= "island", palette = "Dark2")

plt.show()
```

# 3. Strip Plot
A strip plot is used to display individual data points for a categorical variable. It shows how data is distributed across categories.

```python
plt.figure(figsize=(6,3))
sns.set_style("ticks") # used for themes

sns.stripplot(data = penguins, x= 'species', y= "body_mass_g", hue = "island" , dodge= True)
# dodge is a parameter used in categorical plots
# Separated (side-by-side) → dodge=True
# Overlapping (same position) → dodge=False

plt.show()
```

# 4. Histogram
Shows distribution of a single variable.

Data grouped into bins (frequency)

```python
plt.figure(figsize=(6,3))
sns.set_context("notebook")
sns.histplot(x = "body_mass_g", data=penguins,hue="sex", multiple="stack")
plt.show()
```

# 5. Regression Plot
A regression plot shows the relationship between two variables along with a best-fit line (regression line).

It combines:
▫️Scatter plot (data points)
▫️Regression line (trend)


```python
plt.figure(figsize=(6,3))
sns.regplot(x="body_mass_g", y="flipper_length_mm", data=penguins, color = "green" )
plt.show()
```

# 6. Joint Plot
Combines scatter plot + histograms

Shows relationship + distribution together

```python
plt.figure(figsize=(6,3))
sns.jointplot(x="body_mass_g", y="flipper_length_mm", data= penguins)
# sns.jointplot(x="body_mass_g", y="flipper_length_mm", data= penguins,hue='sex')
plt.show()
```

# 7. Bar Plot
Shows comparison between categories

Displays average (mean) values

```python
plt.figure(figsize=(6,3))
sns.barplot(x="body_mass_g", y="species", data= penguins, hue = "sex", palette = ["pink","skyblue"],estimator = np.var)
plt.show()
```

# 8. Count Plot
Displays number of occurrences of each category

Works on categorical data

Similar to a bar chart, but automatically counts values

```python
plt.figure(figsize=(6,3))
sns.set_style('whitegrid')
sns.countplot(data = penguins, x = "species", hue = "island", palette = "Set2")
plt.show()
```

# 9. Box Plot
A box plot shows the distribution of numerical data using quartiles and helps detect outliers.
    
Key Points :-

▫️Shows median & quartiles
▫️Detects outliers easily
▫️Best for numerical + categorical data combination

```python
plt.figure(figsize=(6,3))
sns.boxplot(data = penguins, x= "species", y= 'body_mass_g', hue="sex")
plt.show()
```

# 10. Violin Plot
A violin plot shows the distribution of numerical data and its density (shape).
It is a combination of box plot + KDE (Kernel Density Estimation).
                                       
Key Points :-                                   
▫️Combines box plot + density plot                                     
▫️Shows data shape clearly     
▫️Good for comparing distributions

▪️Width of the violin: Density of the data at different values
▪️Middle line: Median
▪️Thick bar: Interquartile range (if inner: 'box')

```python
plt.figure(figsize=(6,3))
sns.violinplot(data = penguins, x= "species", y= 'body_mass_g', hue="sex", split = True)
plt.show()
```

# 11. KDE Plot
A KDE (Kernel Density Estimation) plot shows the probability density (smooth distribution) of numerical data.

Key Points :-
▫️Smooth version of histogram
▫️Shows density, not count
▫️Good for continuous data
▫️Easy comparison using multiple curves

```python
plt.figure(figsize=(6,3))
sns.kdeplot(data = penguins, x= 'body_mass_g', hue="species", fill = True)
plt.show()
```

# 12. Heatmap
A heatmap is used to visualize data in matrix form using colors.

It helps quickly identify patterns, correlations, and intensity values.

```python
columns =['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', "body_mass_g"]
penguins[columns].corr()
```

```python
plt.figure(figsize=(6,3))
sns.heatmap(data=penguins[columns].corr(), annot=True , cmap="Blues", linewidth ="1",linecolor="black")
# linewidth = outline on box
plt.show()
```

# 13. Rug Plot
A rug plot displays individual data points as small ticks on an axis.

It helps you see the exact distribution of data points.

```python
plt.figure(figsize=(6,3))
sns.rugplot(data = penguins, y = 'body_mass_g' , hue = "species", palette = 'Set2' ,height=0.8)
plt.show()
```

# 14. Pair Plot
A pair plot shows relationships between multiple numerical variables in a dataset.

It creates a grid of plots (scatter plots + hist/KDE on diagonal).

```python
sns.pairplot(data = penguins, hue ='sex' ,palette ='Set2', diag_kind = 'kde')
plt.show()
```

# 15. PairGrid
PairGrid is a more flexible version of pairplot.

It allows you to customize each plot in the grid manually.

```python
graph = sns.PairGrid(data = penguins, hue= "sex", palette = "Set2")

graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)

graph.add_legend()

plt.show()
```

context controls the overall size and scaling of your plot elements like:
▫️text (titles, labels)
▫️lines
▫️markers

"paper" → small (for research papers)

"notebook" → default (balanced)

"talk" → larger (for presentations)
