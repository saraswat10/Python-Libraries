# 10. Concatenate & Merge DataFrame (JOINS)

```python
# Creating DataFrame
df1 = pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
df1
```

```python
# Creating DataFrame
df2 = pd.DataFrame({'ID':[1,2,2,4],'Score':[88,96,77,79]})
df2
```

## Concatenate DataFrames (pd.concat)
Combine objects along a particular axis.

Concept :- 
▫️Just joins DataFrames together 
▫️No key matching
▫️Works row-wise or column-wise

```python
# Concatenate: vertical / row level / top on top

pd.concat([df1, df2], axis=0)
```

```python
# Concatenate: horizontal / column level / side on side

pd.concat([df1, df2], axis=1)
```

## Merge DataFrames (pd.merge)
Performs join operations similar to relational databases like SQL JOINS.

Concept :-
▫️Like SQL JOIN
▫️Combines based on common column(s)

```python
# merge = Join
pd.merge(df1,df2, how = 'inner', on= 'ID')
```

```python
pd.merge(df1, df2, how= 'inner', left_on='ID',right_on='ID')
```

# 13. Lambda Function

Lambda functions in pandas are anonymous one-line functions mainly used with apply, map, and groupby for quick transformations.

Use lambda for :
▫️Simple transformations
▫️Quick conditions
▫️Short logic

Avoid lambda when :
▫️Logic is complex → use def
▫️Needs reuse

```python
# apply ()

df["marks"] = df["marks"].apply(lambda x: x + 5)    # column-wise

df["result"] = df.apply(lambda row: row["marks"] * 2, axis=1)    # row-wise
```

```python
# map()

df["name"] = df["name"].map(lambda x: x.upper())
```

```python
# applymap()

df = df.applymap(lambda x: x * 2)
```

```python
# groupby()

df.groupby("name").agg(lambda x: x.max() - x.min())
```

```python
# Conditional Logic

df["grade"] = df["marks"].apply(
    lambda x: "Pass" if x >= 40 else "Fail"
)
```

NOTE :- all the above lambda methods are not executed 

```python
# creating dataframe

df = pd.DataFrame({
    "name": ["A", "B", "C"],
    "marks": [35, 50, 90]
})
```
