# 5. Array Reshaping and Flattening

##### Reshape Array: changing its dimensions (e.g., rows and columns) without altering the actual data
##### Flatten Array: transforming a multi-dimensional array into a single-dimensional array

```python
arr1= np.array([1,2,3,4,5,6])    # shape = (1,6) - 1 row and 6 columns
print(arr1)
```

```python
# Reshaping
reshaped = arr1.reshape((6,1))    # shape = (6,1) - 6 rows and 1 column
print(reshaped)
```

```python
reshaped2 = arr1.reshape((2,3))     # re-shaped array from (1,6) to (2,3)
print(reshaped2)
```

##### Both convert an array into 1D, but:

##### ravel() → returns a view (shares memory)
##### flatten() → returns a copy (new memory)

```python
#  ravel              convert 1D array
ravel = reshaped2.ravel()
print(ravel)
```

```python
# Flattening
print(reshaped2.flatten())     # flattened array from higher (2D) dimensional (2,3) to (1,6) (1D)
```

# 6. Array Stacking and Splitting

##### Array Stacking: combining multiple arrays along a new axis, resulting in an array with a higher dimension than the input arrays
##### Array Splitting: Divides an array into multiple smaller sub-arrays

```python
# STACKING

a= np.array([1,2,3])
b=np.array([4,5,6])

print(np.vstack((a,b)))  # vertical Stacking - row wise
print()
print(np.hstack((a,b)))  # Horizontal Stacking - column wise
```

```python
c= np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
# SPLITTING

split= np.hsplit(c,3)  # Horizontal split

for s in split:
    print(s)
```

```python
split = np.vsplit(c,2)  # verticall split

for s in split:
    print(s)
```

# 11. Broadcasting

##### Broadcasting is a mechanism that allows arithmetic operations to be performed on arrays of different shapes and sizes, it eliminates the need for explicit loops or reshaping operations, making code more concise, readable and efficient.

##### Broadcasting Rules:
    # Compare shapes from right to left.
    # Dimensions must match or be 1.
    # If one array has fewer dimensions, pad its shape on the left with 1s.
    # Any dimension equal to 1 can stretch to match the other array.

```python
a=np.array([1,2,3,4])
b= np.array([5])
print(a+b)         # adding 1D with 0ne-element array to 1D with multi-element array
```

```python
c=np.array([[1,2,3],[4,5,6]])
d = np.array([10,20,30])
print(c+d)         # adding 2D with 6-element array (2,3) to 1D with 6-element array
```

# 15. Conditional-Based Operations on Array

```python
arr = np.array([1,2,3,4,5])
print(arr)
```

```python
result1 = np.where(arr <2,"low","high")   # WHERE - condition
print(result1)
```

```python
# ArgWhere : Row-Column positions for 2nd arrays

arr= np.array([10,25,30,50])
print(arr)
```

```python
print(np.argwhere(arr>20))    # argwhere - condition
```

```python
mask = np.logical_and(arr>15, arr< 35)       # logical_and  - condition
print(mask)
```
